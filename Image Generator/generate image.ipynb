```python
#pic = load_img('20221231_124559.jpg')
#pic = load_img('20221231_124541.jpg')
pic = load_img('20221231_124530.jpg')
pic_array = img_to_array(pic)
```

```python
pic_array.shape
```

```python
pic_array = pic_array.reshape((1,) + pic_array.shape) # Converting into 4 dimension array
pic_array.shape
```

```python
# Generate 10 images
# batch_size: At a time, how many image should be created.
count = 0
for batch in datagen.flow(pic_array, batch_size=1,save_to_dir="/content/drive/MyDrive/Colab Notebooks/Image", save_prefix='EE', save_format='jpeg'):
    count += 1
    if count > 15:
        break
```
